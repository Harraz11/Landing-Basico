```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np

def draw_plot():

    df = pd.read_csv(
        "epa-sea-level.csv"
    )

    plt.scatter(
        df['Year'],
        df['CSIRO Adjusted Sea Level']
    )

    result = linregress(
        df['Year'],
        df['CSIRO Adjusted Sea Level']
    )

    years_extended = np.arange(
        1880,
        2051
    )

    line = (
        result.slope *
        years_extended
    ) + result.intercept

    plt.plot(
        years_extended,
        line,
        'r'
    )

    df_recent = df[
        df['Year'] >= 2000
    ]

    result_recent = linregress(
        df_recent['Year'],
        df_recent['CSIRO Adjusted Sea Level']
    )

    years_recent = np.arange(
        2000,
        2051
    )

    line_recent = (
        result_recent.slope *
        years_recent
    ) + result_recent.intercept

    plt.plot(
        years_recent,
        line_recent,
        'g'
    )

    plt.xlabel('Year')

    plt.ylabel(
        'Sea Level (inches)'
    )

    plt.title(
        'Rise in Sea Level'
    )

    fig = plt.gcf()

    return fig
```
